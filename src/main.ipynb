```python
import DAPpyr as dap
import matplotlib.pyplot as plt
```

```python
f = dap.EXPT.Expt('enkf_test', {'expt_flag': 0, 'T': 50, 'model_flag': 1, 
                          'Ne': 100, 'dt': 0.05, 'sig_y': 1, 
                          'roi_kf': 0.001, 'gamma':0.0, 'obf': 1, 
                          'doSV':False, 'stepSV' : 3, 'localize': True})
dap.runDA.runDA(f)
plt.plot(f.rmse, label = f.exptname)
plt.legend()
plt.show()
```

```python
f1 = dap.EXPT.Expt('enkf_test', {'expt_flag': 0, 'T': 50, 'model_flag': 1, 
                          'Ne': 1000, 'dt': 0.05, 'sig_y': 1, 
                          'roi_kf': 0.001, 'gamma':0.0, 'obf': 1, 
                          'doSV':False, 'stepSV' : 3, 'localize': False})
f2 = dap.EXPT.Expt('enkf_test', {'expt_flag': 2, 'T': 50, 'model_flag': 1, 
                          'Ne': 1000, 'dt': 0.05, 'sig_y': 1, 
                          'roi_kf': 0.001, 'gamma':0.0, 'obf': 1, 
                          'doSV':False, 'stepSV' : 3, 'localize': False})
f2.copyStates(f1)
```

```python
dap.runDA.runDA(f1)
plt.plot(f1.rmse, label = f1.exptname)
plt.plot(f2.rmse, label = f2.exptname)
plt.legend()
plt.show()
```
